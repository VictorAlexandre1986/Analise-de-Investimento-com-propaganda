# Machine Learning para analisar retorno feito em anuncios TV,Radio,Newspaper

```python
import pandas as pd
```

#### Importando os dados

```python
publi = pd.read_csv('dataset.csv')
```

```python
publi.head()
```

#### Separando os dados x e y

```python
x = publi[['TV','Radio','Newspaper']]
y = publi['Sales']
```

#### Vendo as separações

```python
print(f'Numero de linhas e colunas : {x.shape}')
print(f'Numero de linhas e colunas : {y.shape}')
```

### Explicação do modelo utilizado

#### Modelo de regressão linear será usado ja que se trata de uma previsão quantitativa, queremos entender a relação entre as variaveis 'x investimento em publicidade' e 'y vendas do produto'   

```python
import seaborn as sns
```

```python
#size é o tamanho do gráfico
#kind desenha a linha de regresão
sns.pairplot(publi,x_vars=['TV','Radio','Newspaper'], y_vars=['Sales'], height=7, kind='reg')
```

### Separando os dados para treino e teste

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

### Importação do modelo de regressão linear

```python
from sklearn.linear_model import LinearRegression

#INstanciando o modelo
reglin = LinearRegression()

#Treinando o modelo
reglin.fit(x_train,y_train)
```

```python
#Vendo os coeficientes para TV, Radio e Jornal: aumento nas 
#vendas para cada uma unidade(Real ou Dolar fica seu criterio)
#Por exemplo 1 dolar investido na tv teve vendas de 0.05
#1 dolar investido em Radio teve vendas de 0.19
#Lembrando que esses valores de venda muda a cada divisao de 
#dados em treino e teste
#Zip junta duas listas
list(zip(['TV','Radio','Newspaper'],reglin.coef_))
```

### Previsão

```python
#Esses dados dentro do array foram pegos la no começo onde chama o metodo head
#Pegamos a primeira linha sem o indice e o sales
#É necessário dois colchetes para converter em uma matriz 2D
print(reglin.predict([[230.1,37.8,69.2]]))
#O Resultado obtido está próximo do valor de sales no metodo head
```

```python
#Agora com todos os dados de teste
y_prev = reglin.predict(x_test)
print(y_prev)
```

## Avaliação da performance

### MAE (Mean absolute error)

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test,y_prev))
```

### MSE (Mean squared error)

```python
print(metrics.mean_squared_error(y_test,y_prev))
```

### RMSE (Root mean squared error)

```python
import numpy as np
print(np.sqrt(metrics.mean_squared_error(y_test,y_prev)))
```
